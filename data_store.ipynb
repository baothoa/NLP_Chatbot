```python
import re

def categorize_title(title):
    if not isinstance(title, str):
        return "Không xác định"
    
    title = title.upper()

    if "EVENT CAKE" in title:
        return "Bánh Event sinh nhật"
    
    elif re.search(r"\bSET\b", title):
        return "Set bánh mini Chewy"
    
    else:
        return "Bánh su Medium"
```

```python
# RAW TEXT -> VECTOR -> đưa vào ChromaDB
from sentence_transformers import SentenceTransformer
from typing import List

embedding_model = SentenceTransformer("keepitreal/vietnamese-sbert")

def get_embedding(text: str) -> List[float]:
    if not text or not text.strip():
        print("Empty text!")
        return []

    embedding = embedding_model.encode(text, normalize_embeddings=True)
    return embedding.tolist()
```

```python
df_chewy = df.copy()  # tạo bản sao để không làm thay đổi df gốc

df_chewy["category"] = df_chewy["category"].replace(
    "Bánh Event sinh nhật",
    "Bánh Event Sinh Nhật"
)

df_chewy.head(5)
```

```python
# DataFrame → text → embedding → ChromaDB
import re
import pandas as pd
import chromadb
from chromadb.utils import embedding_functions

# 1. Tạo embedding function

# Dùng model tiếng Việt để biến text -> vector
sentence_transformer_ef = embedding_functions.SentenceTransformerEmbeddingFunction(
    model_name="keepitreal/vietnamese-sbert"
)

# 2. Khởi tạo ChromaDB client

# VECTOR_STORE là thư mục lưu dữ liệu vector xuống ổ cứng
client = chromadb.PersistentClient(path="VECTOR_STORE")

# Nếu muốn xóa dữ liệu cũ (tránh trùng) thì bỏ comment
client.delete_collection("chewy_chewy_aihi_01")
client.delete_collection("chewy_chewy_aihi_02")

# 3. Tạo collection (giống bảng trong DB)

# collection1: dùng title + description
collection1 = client.get_or_create_collection(
    name="chewy_chewy_aihi_01",
    embedding_function=sentence_transformer_ef
)

# collection2: dùng title + category
collection2 = client.get_or_create_collection(
    name="chewy_chewy_aihi_02",
    embedding_function=sentence_transformer_ef
)

# 4. Hàm tiền xử lý text

def preprocess_text(text: str) -> str:
    if pd.isna(text):
        return ""
    text = str(text).lower() # chuyển về chữ thường
    text = re.sub(r"[^\w\s,.]", "", text) # xóa ký tự đặc biệt
    text = re.sub(r"\s+", " ", text).strip() # xóa khoảng trắng thừa
    return text

# 5. Chuẩn hóa dữ liệu sản phẩm

def normalize_product(product: dict) -> dict:
    return {
        "url": "" if pd.isna(product.get("url")) else str(product.get("url")),
        "title": "" if pd.isna(product.get("title_clean")) else str(product.get("title_clean")),
        "description": "" if pd.isna(product.get("description")) else str(product.get("description")),
        "price": "" if pd.isna(product.get("price")) else str(product.get("price")),
        "image_url": "" if pd.isna(product.get("image_url")) else str(product.get("image_url")),
        "category": "" if pd.isna(product.get("category")) else str(product.get("category")),
    }

# 6. Thêm dữ liệu vào collection 1

# Dùng title + description (tìm kiếm theo nội dung)
def add_product_to_chromadb_01(product: dict, collection):
    product = normalize_product(product)

    combined_text = f"{product['title']} {product['description']}"
    combined_text = preprocess_text(combined_text)

    if not combined_text or not product["url"]:
        return

    collection.upsert( # dùng upsert để tránh lỗi trùng ID
        ids=[product["url"]],
        documents=[combined_text],
        metadatas=[{
            "url": product["url"],
            "title": product["title"],
            "description": product["description"],
            "price": product["price"],
            "image_url": product["image_url"],
            "category": product["category"]
        }]
    )

# 7. Thêm dữ liệu vào collection 2

# Dùng title + category (tìm kiếm theo loại)
def add_product_to_chromadb_02(product: dict, collection):
    product = normalize_product(product)

    combined_text = f"{product['title']} {product['category']}"
    combined_text = preprocess_text(combined_text)

    if not combined_text or not product["url"]:
        return

    collection.upsert(
        ids=[product["url"]],
        documents=[combined_text],
        metadatas=[{
            "url": product["url"],
            "title": product["title"],
            "description": product["description"],
            "price": product["price"],
            "image_url": product["image_url"],
            "category": product["category"]
        }]
    )

# 8. Đưa toàn bộ dữ liệu vào ChromaDB

for _, product in df_chewy.iterrows():
    add_product_to_chromadb_01(product.to_dict(), collection1)

for _, product in df_chewy.iterrows():
    add_product_to_chromadb_02(product.to_dict(), collection2)

print("Đã thêm dữ liệu vào ChromaDB thành công.")
```

```python
def weighted_reciprocal_rank(doc_lists, weights=None, c=60):
    """
    Gộp nhiều danh sách kết quả bằng Weighted Reciprocal Rank Fusion (RRF).

    Args:
        doc_lists: List các danh sách document.
        weights: Trọng số cho từng danh sách. Nếu None thì mọi danh sách có weight = 1.
        c: Hằng số trong công thức RRF.

    Returns:
        List document đã được sắp xếp theo điểm RRF giảm dần.
    """
    if weights is None:
        weights = [1] * len(doc_lists)

    if len(doc_lists) != len(weights):
        raise ValueError("Số lượng doc_lists phải bằng số lượng weights.")

    rrf_score_dict = {}
    doc_map = {}

    for doc_list, weight in zip(doc_lists, weights):
        for rank, doc in enumerate(doc_list, start=1):
            doc_id = doc["_id"]
            score = weight * (1 / (rank + c))

            if doc_id not in rrf_score_dict:
                rrf_score_dict[doc_id] = 0.0
                doc_map[doc_id] = doc

            rrf_score_dict[doc_id] += score

    sorted_doc_ids = sorted(
        rrf_score_dict.keys(),
        key=lambda x: rrf_score_dict[x],
        reverse=True
    )

    return [doc_map[doc_id] for doc_id in sorted_doc_ids]


def parse_chroma_results(results):
    """
    Chuyển kết quả từ ChromaDB thành list dict dễ xử lý hơn.
    """
    parsed = []

    if not results["ids"] or not results["ids"][0]:
        return parsed

    for i in range(len(results["ids"][0])):
        parsed.append({
            "_id": results["ids"][0][i],
            "url": results["metadatas"][0][i].get("url", ""),
            "title": results["metadatas"][0][i].get("title", ""),
            "description": results["metadatas"][0][i].get("description", ""),
            "price": results["metadatas"][0][i].get("price", ""),
            "image_url": results["metadatas"][0][i].get("image_url", ""),
            "category": results["metadatas"][0][i].get("category", ""),
            "distance": results["distances"][0][i] if "distances" in results else None
        })

    return parsed


def hybrid_search(query: str, collection, limit=10):
    """
    Hybrid search bằng 2 lượt truy vấn:
    1. Query gốc
    2. Query đã preprocess

    Sau đó gộp kết quả bằng RRF.
    """
    if not query or not query.strip():
        return []

    clean_query = preprocess_text(query)

    results_1 = collection.query(
        query_texts=[query],
        n_results=limit
    )

    results_2 = collection.query(
        query_texts=[clean_query],
        n_results=limit
    )

    parsed_results_1 = parse_chroma_results(results_1)
    parsed_results_2 = parse_chroma_results(results_2)

    fused_documents = weighted_reciprocal_rank(
        doc_lists=[parsed_results_1, parsed_results_2],
        weights=[1, 1]
    )

    return fused_documents[:limit]


def enhance_prompt(query: str, collection, limit=10):
    """
    Tạo context/prompt từ kết quả hybrid search.
    """
    knowledge = hybrid_search(query, collection, limit)

    if not knowledge:
        return "Không tìm thấy sản phẩm phù hợp."

    enhanced_prompt = ""
    for i, result in enumerate(knowledge, start=1):
        enhanced_prompt += (
            f"[Sản phẩm {i}]\n"
            f"Title: {result.get('title', 'N/A')}\n"
            f"Description: {result.get('description', 'N/A')}\n"
            f"Price: {result.get('price', 'N/A')}\n"
            f"Image URL: {result.get('image_url', 'N/A')}\n"
            f"URL: {result.get('url', 'N/A')}\n"
            f"Category: {result.get('category', 'N/A')}\n\n"
        )

    return enhanced_prompt
```

```python
def display_result(results: list):
    if not results:
        print("Không có kết quả.")
        return

    for i, product in enumerate(results, start=1):
        print(f"[Sản phẩm {i}]")
        print("Tên:", product.get("title", "N/A"))
        print("Giá:", product.get("price", "N/A"))
        print("Mô tả:", product.get("description", "N/A"))
        print("Hình ảnh:", product.get("image_url", "N/A"))
        print("Danh mục:", product.get("category", "N/A"))
        print("Link:", product.get("url", "N/A"))
        print("-" * 40)
```
